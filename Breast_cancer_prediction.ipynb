```python
# Take the dependent and independent feature for prediction
prediction_feature = [ "radius_mean",  'perimeter_mean', 'area_mean', 'symmetry_mean', 'compactness_mean', 'concave points_mean']

targeted_feature = 'diagnosis'

len(prediction_feature)
```

```python
X = data[prediction_feature]
X

print(X.shape)
print(X.values)
```

```python
y = data.diagnosis
y

print(y.values)
```

```python
# Split the dataset into TrainingSet and TestingSet by 33% and set the 15 fixed records

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=15)

print(X_train)
print(X_test)
```

```python
# Perform Feature Standerd Scalling
# Standardize features by removing the mean and scaling to unit variance

# The standard score of a sample x:
# scaled value = (x - u) / s

# Scale the data to keep all the values in the same magnitude of 0 -1 
# ensure that all features have the same scal
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# ML Model Selecting and Model PredPrediction

# Model Building

#  function for:
# model building,
# preforming prediction
# measure it's prediction and accuracy score.


def model_building(model, X_train, X_test, y_train, y_test):
    """
    Model Fitting, Prediction, and Evaluation
    Returns: (score, accuracy_score, f1_score, confusion_matrix, predictions)
    """
    
    model.fit(X_train, y_train)
    score = model.score(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy = accuracy_score(predictions, y_test)
    f1 = f1_score(y_test, predictions)
    cm = confusion_matrix(y_test, predictions)
    
    return (score, accuracy, f1, cm, predictions)

```

```python
# a dictionary for multiple models for bulk predictions

models_list = {
    "LogisticRegression" :  LogisticRegression(),
    "RandomForestClassifier" :  RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=5),
    "DecisionTreeClassifier" :  DecisionTreeClassifier(criterion='entropy', random_state=0),
    "SVC" :  SVC(),
}

print(models_list)
```

```python
# Loop through models and evaluate them
for model_name, model in models_list.items():
    print(f"Evaluating {model_name}...")
    score, accuracy, f1, cm, predictions = model_building(model, X_train, X_test, y_train, y_test)
    print(f"Model: {model_name}")
    print(f"Training Score: {score}")
    print(f"Test Accuracy: {accuracy}")
    print(f"F1 Score: {f1}")
    print(f"Confusion Matrix:\n{cm}\n")
```

```python
# check the key and values to store it's values in DataFrame below.

print(list(models_list.keys()))
print(list(models_list.values()))

print(zip(list(models_list.keys()), list(models_list.values())))
```

```python
# Model implementation
# Training model while showing classification report particular model

# function for confusion metric Graphs

def cm_metrix_graph(cm):
    
    sns.heatmap(cm,annot=True,fmt="d")
    plt.show()
```

```python
# sending categorical features for to the prediction
df_prediction = []
confusion_matrixs = []
df_prediction_cols = [ 'model_name', 'score', 'accuracy_score' , "accuracy_percentage"]

for name, model in zip(list(models_list.keys()), list(models_list.values())):
    
    (score, accuracy, predictions) = model_building(model, X_train, X_test, y_train, y_test )
    
    print("\n\nClassification Report of '"+ str(name), "'\n")
    
    print(classification_report(y_test, predictions))

    df_prediction.append([name, score, accuracy, "{0:.2%}".format(accuracy)])
    
    # For Showing Metrics
    confusion_matrixs.append(confusion_matrix(y_test, predictions))
    
        
df_pred = pd.DataFrame(df_prediction, columns=df_prediction_cols)
```

```python
def cross_val_scoring(model, data, prediction_feature, targeted_feature):
    # Fit model on the full dataset
    model.fit(data[prediction_feature], data[targeted_feature])
    
    predictions = model.predict(data[prediction_feature])    
    accuracy = accuracy_score(data[targeted_feature], predictions)
    print("\nFull-Data Accuracy:", round(accuracy, 2))

    # Initialize K folds
    kFold = KFold(n_splits=5)  # Define 5 different data folds
    
    err = []  # List to store fold accuracies
    # to store individual fold accuracies
    fold_accuracies = []
    
    for train_index, test_index in kFold.split(data):
        # Data splitting via fold indexes
        X_train = data[prediction_feature].iloc[train_index, :]
        y_train = data[targeted_feature].iloc[train_index]
        
        X_test = data[prediction_feature].iloc[test_index, :]
        y_test = data[targeted_feature].iloc[test_index]
        
        # Train the model
        model.fit(X_train, y_train)
        
        # Predict and calculate accuracy on the test fold
        predictions = model.predict(X_test)
        fold_accuracy = accuracy_score(y_test, predictions)
        fold_accuracies.append(fold_accuracy)
        err.append(fold_accuracy)

    # Print the individual fold scores
    print(f"Cross Validation Scores (5 folds): {', '.join([str(round(score, 2)) for score in fold_accuracies])}")
        
    # Print average accuracy across all folds
    avg_accuracy = np.mean(err)
    print(f"Cross Validation Average Accuracy: {round(avg_accuracy, 2)}")
```

```python
# HyperTunning the ML Model
# Applying tuning parameters

# use GridSearchCV to know the best performing parameters

from  sklearn.model_selection import GridSearchCV
```

```python

# GridSearchCV implements a â€œfitâ€ and a â€œscoreâ€ method. 
# It also implements â€œpredictâ€, â€œpredict_probaâ€, â€œdecision_functionâ€, â€œtransformâ€ 
# and â€œinverse_transformâ€ if they are implemented in the estimator used.

# The parameters of the estimator used to apply these methods are 
# optimized by cross-validated grid-search over a parameter grid.

# Pick the model
model = DecisionTreeClassifier()

# Tunning Params
param_grid = {'max_features': ['auto', 'sqrt', 'log2'],
              'min_samples_split': [2,3,4,5,6,7,8,9,10], 
              'min_samples_leaf':[2,3,4,5,6,7,8,9,10] }


# Implement GridSearchCV
gsc = GridSearchCV(model, param_grid, cv=10) # For 10 Cross-Validation

gsc.fit(X_train, y_train) # Model Fitting

print("\n Best Score is ")
print(gsc.best_score_)

print("\n Best Estinator is ")
print(gsc.best_estimator_)

print("\n Best Parametes are")
print(gsc.best_params_)
```
